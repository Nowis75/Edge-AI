# Klasifikace

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x_red = np.random.randn(50, 2) + [2, 2]
x_blue = np.random.randn(50, 2) + [-2, -2]
# Nastavení stylu

plt.rcParams.update({
    'font.size': 16,
    'axes.titlesize': 14,
    'axes.labelsize': 10,
    'legend.fontsize': 10,
    'xtick.labelsize': 9,
    'ytick.labelsize': 9,
    'lines.linewidth': 1,
    'font.family': 'DejaVu Sans',
})
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(6, 5))
plt.scatter(x_red[:, 0], x_red[:, 1], color='darkblue', label='Třída 1')
plt.scatter(x_blue[:, 0], x_blue[:, 1], color='red',  label='Třída 2')
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
#plt.title('Klasifikace - oddělení tříd')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend(frameon=False)
plt.grid(alpha=0.3)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# Regrese

```python
np.random.seed(0)
x = np.linspace(0, 10, 50)
y = 2.5 * x + np.random.randn(50) * 5

coef = np.polyfit(x, y, 1)
poly1d_fn = np.poly1d(coef)

plt.rcParams.update({
    'font.size': 16,
    'axes.titlesize': 14,
    'axes.labelsize': 10,
    'legend.fontsize': 10,
    'xtick.labelsize': 9,
    'ytick.labelsize': 9,
    'lines.linewidth': 1,
    'font.family': 'DejaVu Sans',
})
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(6, 5))
plt.scatter(x, y, color='darkblue', label='Data')
plt.plot(x, poly1d_fn(x), color='red', label='Lineární regrese')
#plt.title('Regrese - aproximace dat')
plt.xlabel('Vstupní hodnota')
plt.ylabel('Predikovaná hodnota')
plt.legend()
plt.grid(True)
plt.show()
```

# Detekce objektů

```python
import matplotlib.patches as patches

plt.rcParams.update({
    'font.size': 16,
    'axes.titlesize': 14,
    'axes.labelsize': 10,
    'legend.fontsize': 10,
    'xtick.labelsize': 9,
    'ytick.labelsize': 9,
    'lines.linewidth': 1,
    'font.family': 'DejaVu Sans',
})
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(6, 5))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
#ax.set_title('Detekce objektů')

box1 = patches.Rectangle((2, 2), 3, 3, linewidth=2, edgecolor='red', facecolor='none')
box2 = patches.Rectangle((5, 5), 3, 3, linewidth=2, edgecolor='darkblue', facecolor='none')

ax.add_patch(box1)
ax.text(2, 5.5, 'Kostka', color='red')
ax.add_patch(box2)
ax.text(5, 8.5, 'Míč', color='blue')

ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.grid(True)
plt.show()
```

# Segmentace

```python
image = np.random.rand(10, 10)
mask = np.zeros_like(image)
mask[2:7, 2:7] = 1  # Vysegmentovaná oblast

plt.rcParams.update({
    'font.size': 16,
    'axes.titlesize': 14,
    'axes.labelsize': 10,
    'legend.fontsize': 10,
    'xtick.labelsize': 9,
    'ytick.labelsize': 9,
    'lines.linewidth': 1,
    'font.family': 'DejaVu Sans',
})
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(6, 5))
plt.imshow(image, cmap='gray', interpolation='none')
plt.imshow(mask, cmap='jet', alpha=0.5, interpolation='none')
plt.title('Segmentace - zvýraznění oblasti')
plt.axis('off')
plt.show()
```

# Detekce anomálií

```python
normal = np.random.randn(100, 2)
anomalies = np.random.uniform(low=-6, high=6, size=(5, 2))

plt.rcParams.update({
    'font.size': 16,
    'axes.titlesize': 14,
    'axes.labelsize': 10,
    'legend.fontsize': 10,
    'xtick.labelsize': 9,
    'ytick.labelsize': 9,
    'lines.linewidth': 1,
    'font.family': 'DejaVu Sans',
})
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(6, 5))
plt.scatter(normal[:, 0], normal[:, 1], label='Normální data', color='darkblue')
plt.scatter(anomalies[:, 0], anomalies[:, 1], label='Anomálie', color='red', marker='x', s=100)
#lt.title('Detekce anomálií - odlehlé body')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
