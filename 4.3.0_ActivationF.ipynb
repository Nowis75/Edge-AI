## Aktivační funkce

```python

import numpy as np
import matplotlib.pyplot as plt

# Definice aktivačních funkcí
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

# Vstupní rozsah pro zobrazení funkcí
x = np.linspace(-10, 10, 400)

# Výpočty výstupů
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)

# Vykreslení grafů
plt.figure(figsize=(22, 4))

plt.subplot(1, 3, 1)
plt.plot(x, y_sigmoid, label="Sigmoid", color='blue')
plt.title("Sigmoid funkce")
plt.grid(True)
plt.xlabel("x")
plt.ylabel("σ(x)")

plt.show()
```

```python
plt.figure(figsize=(22, 4))

plt.subplot(1, 3, 2)
plt.plot(x, y_tanh, label="Tanh", color='green')
plt.title("Tanh funkce")
plt.grid(True)
plt.xlabel("x")
plt.ylabel("tanh(x)")


plt.show()
```

```python
plt.figure(figsize=(22, 4))
plt.subplot(1, 3, 3)
plt.plot(x, y_relu, label="ReLU", color='red')
plt.title("ReLU funkce")
plt.grid(True)
plt.xlabel("x")
plt.ylabel("ReLU(x)")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    e_x = np.exp(x - np.max(x, axis=0, keepdims=True))  # stabilizace
    return e_x / np.sum(e_x, axis=0, keepdims=True)

# Simulace: měníme jednu hodnotu a sledujeme vliv na výstup softmax
x_vals = np.linspace(-5, 5, 300)
outputs = []

# Dvě složky fixní, jen jedna mění svou hodnotu (např. třída 1)
for x in x_vals:
    vec = np.array([x, 0.0, -1.0])
    out = softmax(vec)
    outputs.append(out)

outputs = np.array(outputs)

# Vizualizace jednotlivých výstupů softmax (třídy)
plt.figure(figsize=(16, 4))
plt.subplot(1, 3, 3)
plt.plot(x_vals, outputs[:, 0], label='Třída 1 (mění se)', linewidth=2)
#plt.plot(x_vals, outputs[:, 1], label='Třída 2 (fixní = 0)', linestyle='--')
#plt.plot(x_vals, outputs[:, 2], label='Třída 3 (fixní = -1)', linestyle=':')
plt.title("Softmax funkce")
plt.xlabel("x")
plt.ylabel("Pravděpodobnost podle softmax")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
